```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import numpy as np

diskexample1 = Ellipse((0,0),3,3 * np.cos(np.deg2rad(30.)),angle=0)
ax = plt.subplot(111, aspect='equal')
ax.add_patch(diskexample1)
ax.set_xlim(left = 2, right=-2)
ax.set_ylim(bottom=-2, top=2)
plt.show()
```

```python
diskexample2 = Ellipse((0,0),1.5,1.5 * np.cos(np.deg2rad(30.)),angle=90-15)
ax = plt.subplot(111, aspect='equal')
ax.add_patch(diskexample2)
ax.plot(np.linspace(0,1),np.linspace(0,1)*np.cos(np.deg2rad(30.)),color='k' )
ax.set_xlim(left = 1, right=-1)
ax.set_ylim(bottom=-1, top=1)
ax.vlines(0, 0, 1, color='k')
plt.show()
```

```python
diskexample3 = Ellipse((0,0),1 / np.cos(np.deg2rad(50.)),1,angle=90-120)
ax = plt.subplot(111, aspect='equal')
ax.add_patch(diskexample3)
ax.plot(np.linspace(0,1),np.linspace(0,1)*np.cos(np.deg2rad(30.)),color='k' )
ax.set_xlim(left = 1, right=-1)
ax.set_ylim(bottom=-1, top=1)
ax.vlines(0, 0, 1, color='k')
plt.show()
```

```python
from astropy.io import fits
import numpy as np

HD163296_continuum_file = "/Users/mengfeilong/desktop/HD163296_continuum.fits"
HD163296_continuumcube = fits.open(HD163296_continuum_file)[0]
header = HD163296_continuumcube.header
data = HD163296_continuumcube.data
data = np.squeeze(data)
naxis1 = header['NAXIS1']
naxis2 = header['NAXIS2']
cdelt1 = header['CDELT1']*3600
cdelt2 = header['CDELT2']*3600
ax=plt.subplot(111,aspect='equal')
ax.imshow(data[1000:-1000,1000:-1000], 'magma', origin='lower', extent = [-0.5*1/3*cdelt1*naxis1, 0.5*1/3*cdelt1*naxis1,-0.5*1/3*cdelt2*naxis2, 0.5*1/3*cdelt2*naxis2 ])
ax.set_xlabel('x (arcseconds)')
ax.set_ylabel('y (arcseconds)')
plt.show()
```
